# Bar Chart (Average Rating by Airline)
# Data Prep: Group by airline and calculate the mean rating

```python
# Convert the series to a DataFrame for cleaner seaborn usage
avg_rating_per_airline = df.groupby('AirLine_Name')['Rating - 10'].mean().sort_values(ascending=False)
avg_rating_df = avg_rating_per_airline.reset_index(name='Average Rating')

plt.figure(figsize=(10, 6))

# FIX: Assign 'AirLine_Name' to the hue parameter and set legend=False
sns.barplot(x='AirLine_Name', 
            y='Average Rating', 
            data=avg_rating_df, 
            palette='viridis',
            hue='AirLine_Name', 
            legend=False)
            
plt.title('Average Rating by Airline', fontsize=14)
plt.xlabel('Airline Name', fontsize=12)
plt.ylabel('Average Rating (Out of 10)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

# Line Chart (Monthly Review Volume Over Time)
# Data Prep: Count reviews by year-month period

```python
monthly_reviews = df['Date_Parsed'].dt.to_period('M').value_counts().sort_index()
monthly_reviews_df = monthly_reviews.rename_axis('Month').reset_index(name='Count')
# Convert Period to string for plotting on the x-axis
monthly_reviews_df['Month'] = monthly_reviews_df['Month'].astype(str)

plt.figure(figsize=(12, 6))
plt.plot(monthly_reviews_df['Month'], monthly_reviews_df['Count'], marker='o', linestyle='-', color='teal')
plt.title('Monthly Review Volume Over Time', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Number of Reviews', fontsize=12)
# Show every 6th month tick for readability
plt.xticks(monthly_reviews_df['Month'][::6], rotation=45, ha='right')
plt.grid(axis='both', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

# Pie Chart (Proportion of Verified Reviews)
# Data Prep: Check for Trip Verified' status

```python
verified_status = df['Review'].str.contains('✅ Trip Verified', na=False).value_counts()
verified_status.index = ['Verified' if x else 'Not Verified' for x in verified_status.index]

plt.figure(figsize=(8, 8))
plt.pie(verified_status.values, labels=verified_status.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Proportion of Verified vs. Non-Verified Reviews', fontsize=14)
plt.tight_layout()
plt.show()
```

#  Data Preparation: Create the missing column ---
# Calculates the length of the text in the 'Review' column and stores it in the new 'Review_Length' column.
# Plotting the Scatter Plot (Now it will work) ---
# Drop rows where Rating or Review_Length might be missing for a clean plot

```python

df['Review_Length'] = df['Review'].str.len()

scatter_data = df.dropna(subset=['Rating - 10', 'Review_Length'])
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Rating - 10', y='Review_Length', data=scatter_data, alpha=0.6, color='darkorange')
plt.title('Review Length vs. Customer Rating', fontsize=14)
plt.xlabel('Rating (Out of 10)', fontsize=12)
plt.ylabel('Review Length (Characters)', fontsize=12)
plt.grid(axis='both', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

```

# --. Data Preparation: Create the missing column ---
# Convert the original 'Date' column (object type) to datetime objects
# --. Heatmap Plotting (Now it will work) ---
# Group by month and recommendation status, then unstack to create a matrix

```python

df['Date_Parsed'] = pd.to_datetime(df['Date'], errors='coerce')

recommendation_by_time = df.dropna(subset=['Date_Parsed']).groupby([
    df['Date_Parsed'].dt.to_period('M').astype(str),
    'Recommond'
]).size().unstack(fill_value=0)

plt.figure(figsize=(14, 8))
# Transpose the data (T) so 'yes'/'no' are on the Y-axis (easier to read)
sns.heatmap(recommendation_by_time.T, cmap='YlGnBu', annot=True, fmt='d', linewidths=.5, cbar_kws={'label': 'Number of Reviews'})
plt.title('Heatmap of Recommendation Count Over Time (Month)', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Recommendation Status', fontsize=12)
plt.yticks(rotation=0)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

# Ensure date is parsed (Fixes the recurring KeyError if not run previously)
# Calculate monthly 'yes' vs. total.Calculate the Recommendation Rate (RR = Yes / (Yes + No)) with zero division handling
# Show the last 6 periods for current trend focus

```python
df['Date_Parsed'] = pd.to_datetime(df['Date'], errors='coerce')

monthly_sentiment = df.dropna(subset=['Date_Parsed']).groupby(df['Date_Parsed'].dt.to_period('M'))['Recommond'].value_counts().unstack(fill_value=0)

monthly_sentiment['Recommendation Rate'] = monthly_sentiment.apply(
    lambda row: (row['yes'] / (row['yes'] + row['no'])) * 100 if (row['yes'] + row['no']) > 0 else np.nan,
    axis=1
)

print("\n--- Monthly Recommendation Rate Trend (Last 6 Months) ---\n")
print(monthly_sentiment['Recommendation Rate'].tail(6))
```

# Heatmap Code.# Aggregate data by month and recommendation status
# Transposing the data (.T) makes 'Recommond' the Y-axis and Months the X-axis for better readability
